## Question 2

#### Imagine that you are responsible to monitor a set of ETL pipelines, which load structured time series data from external sources into the cloud. All of this data is needed as input for ML models, that are running in production.

#### Which KPI's would be most prominent to be closely monitored?

I assume that in order to monitor the ETL pipelines and prevent critical data issue to affect the model: integrity, consistency, and availability of the data are three aspects to consider.

In this case we have the data coming from external sources, so there's the risk that schema or pattern of the data may change over time without notification or the team’s awareness. Data vandalism from an external party or even Software updates in our  infrastructure may severely compromise the integrity of data or event make it unvailable.
The data might also be sudenly unavailable and to monitor that some of the important KPI's might be:
        *Amount of delay time between each batch of data
        *Total minutes, hours, or days, without data in the pipeline
        *Number of times that data is delivered behind schedule

The integrity of the data can be monitored by creating comprehensive validation rules to catch and report the exceptions during the  transformation phase of the ETL process , some of them can be the number of missing or unseen features or ratio of data points that fail each validation rule.

Concerning the consistency, I am thinking that the data, in this case power usage might suddenly change  (due to let's say power outage/blakout..) and the previously trained model would not be suitable to make predictions based on these new data pattern. So as a data scientist/enginner it would be good to have some kind of a dashboard where the data distribution( mean and standard deviation) can be displayed so we can compare the skewness of the new data distribution.

#### What would you do if one of those ETL pipelines starts to fail?

The first thing I guess is to report the failure to the team because it makes a difference when many minds are working to fix something

If the pipeline starts to fail, I would try figure out the origin of the failure then start controlling the damage.
In case of a data consistency issue, I assume that we should refit the model to the new data.
In case of a data intergrity issue, I assume that we should create more rebust data validation rules
In case of a data availability issue, I assume that we should first check wether the problem is internal or external and then fix the errors/ make a patches to make sure this doesn't happen again.


#### What measures could be implemented to make those ETL pipelines as robust as posssible?

Aside from the measures mentioned above which are more oriented toward maintainability  and operational stability, I assume that to maximise the robustness of the data pipeline we'll need to worry about scalability. Data volume might vary throughout the day. Adopting a solution that can scale may avoid a crash in the pipeline
We also need to test to ensure that production code behaves in the way we expect and to to plot as many important metrics as possible on a dashboard and check that dashboard once every day. (This can be done as a team exercise during agile standups).
We finally need to make sure that everyone on our team knows what to do in the event of a failure.
